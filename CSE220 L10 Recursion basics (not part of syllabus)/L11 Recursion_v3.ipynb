### Example 04: Factorial calculation

```python
def fact(n):
    result = 1
    while n > 0:
        result *= n
        n -= 1
    return result

print(fact(4))
```

### Example 05: Fibonachi series summation

```python
def get_nth_fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    elif n > 1:
        n1 = 0
        n2 = 1

        for _ in range(2, n + 1):
            fib_sum = n1 + n2
            n1 = n2
            n2 = fib_sum

        return n2
    else:
        return -1

# Example usage
n = 4
print(f"{n}th Fibonacci number is: {get_nth_fibonacci(n)}")

n = 6
print(f"{n}th Fibonacci number is: {get_nth_fibonacci(n)}")

n = 8
print(f"{n}th Fibonacci number is: {get_nth_fibonacci(n)}")
```

```python
def get_nth_fibonacci_recursive(n):

    if n == 0:
        return 0
    elif n == 1:
        return 1
    elif n>1:
        return get_nth_fibonacci_recursive(n - 1) + get_nth_fibonacci_recursive(n - 2)

n = 4
print(f"{n}th Fibonacci numbers is: {get_nth_fibonacci_recursive(n)}")

n = 6
print(f"{n}th Fibonacci numbers is: {get_nth_fibonacci_recursive(n)}")

n = 8
print(f"{n}th Fibonacci numbers is: {get_nth_fibonacci_recursive(n)}")
```

### Example 06: FW printing of array elements

```python
import numpy as np
arr1D = np.array([12, 44, 6, 57, 69, 87, 5, 3])

i = 0
while i< len(arr1D):
    print(arr1D[i],end=" ")
    i+=1
```

```python
def arrForwardPrinting(i, arr):
    if i==len(arr):# Base case #Stops, when i reaches length of array. Making sure does not go Out of bounds
        return
    else:
        print(arr[i], end=' ')
        arrForwardPrinting(i+1, arr) # recursive call

arr1D = np.array([12, 44, 6, 57, 69, 87, 5, 3])
arrForwardPrinting(0, arr1D)
```

### Example 08: Summation of array elements

```python
# USING ITERATIVE PROCESS

def arrSum(arr):
    total = 0
    i = 0
    while i< len(arr):
        total+=arr[i]
        i = i + 1
    return total

arr1D = np.array([10, 20, 30])
result = arrSum(arr1D)
print(result)
```

```python
## USING RECURSION
def arrSum(i, arr):
    # The base case is triggered when i reaches the last index of the array (len(arr) - 1).
    if i==len(arr)-1: # Base case
        return arr[i]
    else:
        return arr[i] + arrSum(i+1, arr) # Recursive call # increasing i by 1

arr1D = np.array([1,4,6,5,9,3])
sum = arrSum(0, arr1D)
print(sum)
```

## 3) Recursion on Linked List

```python
class Node:
    def __init__(self,e,n=None):
        self.elem = e
        self.next = n
```

### Example 11: convert array to linked list

```python
def printLL(head):
    temp = head # i=0
    while temp!=None: # while i< len:
        print(temp.elem, end=' -> ')
        temp = temp.next # i = i+1
    print() # since using end, for going to the next line after end usage

#This function converts array to linked list
def arr2LL(arr):
    head = Node(arr[0])
    current = head # here the current is working as a tail
    i = 1
    while i<len(arr):
        new_node = Node(arr[i]) # append after tail
        current.next = new_node
        current = current.next
        i+=1

    return head

head = arr2LL([10, 20, 30, 40, 50])
printLL(head)
```

```python
def arr2LL_recursive(arr, index=0):
    # Base case: If we've processed all elements, (end of the list)
    if index == len(arr):
        return

    # Create the current node with the element at the current index
    current_node = Node(arr[index])

    # Recursively set the next node for the current node
    current_node.next = arr2LL_recursive(arr, index + 1)

    return current_node  # Return the current node

head = arr2LL_recursive([10, 20, 30, 40, 50])
printLL(head)
```

```python
def find_max_array(arr):
    if len(arr) == 0:
        return None

    max_value = arr[0]
    i = 1
    while i < len(arr):
        if arr[i] > max_value:
            max_value = arr[i]
        i = i + 1

    return max_value

a1 = [10, 20, 5, 15]
print(a1)

maximum = find_max_array(a1)
print(f"The maximum value in the linked list is: {maximum}")
```

```python
class Node:
    def __init__(self, elem, next=None):
        self.elem = elem
        self.next = next

def find_max(head):
    if head is None:
        return None  # Handle the case of an empty list

    # Initialize the maximum value with the data of the head node
    max_value = head.elem# i = 0
    current = head.next  # Start traversing from the second node i= 1

    while current is not None:
        if current.elem > max_value:  # Update max_value if a larger value is found
            max_value = current.elem
        current = current.next  # Move to the next node i = i +1

    return max_value


head = arr2LL([10, 20, 5, 15])
printLL(head)

# Find the maximum value in the linked list
maximum = find_max(head)
print(f"The maximum value in the linked list is: {maximum}")
```

```python
def find_max_recursive(temp):
    # Base case: if there's only one node, return its value
    if temp.next is None:
        return temp.elem

    # Recursive case: find the maximum in the rest of the list
    max_val = temp.elem #  value of the head/temp
    max_node_rest_val = find_max_recursive(temp.next)

    # Compare the current node's value (temp.elem) with the maximum value of the rest
    if max_node_rest_val > max_val:
        max_val = max_node_rest_val
    return max_val


head = arr2LL([10, 20, 5, 25, 3])
printLL(head)

# Find the maximum value in the linked list
maximum = find_max_recursive(head)
print(f"The maximum  value in the linked list is: {maximum }")
```
