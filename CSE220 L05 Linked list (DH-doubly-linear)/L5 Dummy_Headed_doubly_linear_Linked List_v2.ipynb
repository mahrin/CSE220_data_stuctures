## Defining the Node class for doubly linear linked list

````python
# Defining the node class
class Node:
    def __init__(self, elem, next=None, prev=None):
        self.elem = elem # The element that we want to put in the list
        self.next = next # The reference to the next node in the list
        self.prev = prev # The reference to the previous node in the list
````

## 2.5.1 LL Creation:
It is similar to creating a non-dummy singly linaer linked list. The difference is that we need to ensure the connections in a proper way.
The below function takes an array and creates a dummy-headed doubly linear linked list.
- The code is almost same as linear non-dummy singly linked list.
- **Exception--array values appending starts from after the head node**

DHLLL_01_creation.jpg

````python
import numpy as np

def create_doubly_linear_LL(arr):
    head = Node(None)  # Dummy head node

    tail = head # append
    for i in range(0, len(arr), 1):
        #-----append code------------
        newNode = Node(arr[i])

        tail.next = newNode  # Link tail to new node
        newNode.prev = tail  # Link new node back to tail---

        tail = newNode  # Move to the newnode ,# tail = tail.next#---BOTH FINE
        #---------------------------------

    return head  # Returning the head of the newly created linked list
````

## 2.5.2 Iteration (FW & BW):
- It is similar to the previous linked list (non-dummy linear singly ll).

- The difference is that it will start from the next item of the
dummy head and run till the Node with None/Null in the next (tail node).

DHLLL_01 - Copy.jpg

````python
# Function to print the linked list in forward direction
def printLL_forward(head):
    print("Forward Printing: ", end='')
    current_node = head.next  # Start from the first real element
    while current_node != None:
        print(current_node.elem, end=' -> ')
        current_node = current_node.next # i = i + 1
    print("END")  # End of the list

arr = np.array([10, 20, 30, 40, 50])
head3 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_forward(head3)
print('-'*50)

arr = np.array([10, 20, -30])
head4 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_forward(head4)
````

### getTail()

````python
def get_tail(head):
    current_node = head # dummy head will be skipped, if we dont start from head
    # for, size=0, the dummy head is the tail Node.
    while current_node.next!= None:
        current_node = current_node.next
    return current_node
````

````python
# Function to print the linked list in backward direction
def printLL_backward(head):
    print("Backward Printing: ", end='')

    temp_tail = get_tail(head)

    # Now traverse backwards
    while temp_tail.prev != None: #while temp_tail!= head:--BOTH FINE
        print(temp_tail.elem, end=' <- ')
        temp_tail = temp_tail.prev # i = i-`1
    print("HEAD")  # Indicating the start of the list

arr = np.array([10, 20, 30, 40, 50])
head3 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_backward(head3)
print('-'*50)

arr = np.array([10, 20, -30])
head4 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_backward(head4)
print('-'*50)

arr = np.array([])
head5 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_backward(head5)
````

## 2.5.3 Size:
- In order to count the number of nodes present in a linked list, we just need to count how many times the iteration process is running during a traversal.
- **ignore the dummy head, while counting the nodes**
- Forward or Backward traversal--does not matter

DHLLL_02 - Copy.jpg

````python
def get_size(head):
    # If the list is empty (head.next== None), return 0
    if head.next== None:
        return 0

    count = 0
    current_node = head.next
    while current_node!= None:
        count += 1
        current_node = current_node.next
    return count

arr = np.array([])
head5 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_forward(head5)
print("Size:", get_size(head5))
print('-'*50)

arr = np.array([10, 20, 30, 40, 50])
head3 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_forward(head3)
print("Size:", get_size(head3))
print('-'*50)

arr = np.array([10, 20, -30])
head4 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_forward(head4)
print("Size:", get_size(head4))
print('-'*50)
````

## 2.5.4 getNode()/nodeAt():
- Often referred as nodeAt() in few materials/content.
- The getNode method is similar to the previous lecture. **The difference is that the dummy head does not represent any index as it does not have any value.**

DHLLL_02 - Copy.jpg

````python
def getNode(head, idx):
    count = 0
    current_node = head.next  # Start from the first real element after head
    while current_node != None:
        if count == idx:
            return current_node  # Return the node at the given index
        count += 1
        current_node = current_node.next
    return None  # Return None if index is out of bounds

               # 0   1   2   3   4
arr = np.array([10, 20, 30, 40, 50])
head3 = create_doubly_linear_LL(arr)

# printing the newly created linked list
printLL_backward(head3)
print('-'*50)

# Fetching node at index 2
node0 = getNode(head3, 0)
print("Node at index 0 (getting 10, just after Dhead):", node0.elem)


# Fetching node at index 2
node1 = getNode(head3, 2)
print("Node at index 2 (getting 30):", node1.elem)


# Fetching node at index 4
node2 = getNode(head3, 4)
print("Node at index 4 (getting 50, tail):", node2.elem)


# Fetching node at index 5 (which is out of bounds for a list of size 5)
node3 = getNode(head3, 5)
print("Node at index 5 (it does not exist, so None):", node3)
````

## 2.5.5 Insertion:
- To insert a new node in the list, you need the reference to the predecessor to link in the new node.
- Unlike for a singly-linked linear list, **there is no "special" case here**, since there is always a valid predecessor node available, thanks to the dummy head.
- **Head does not need to change.**

### Illustration of insertion after head (index=0) [Special case, **pred=head (pred does not have index--cannot use getNode()**)]

### Illustration of insertion after head.next and before end of LL (generic case)

### Illustration of insertion after the tail (Insert at size) (Special case, Succesor is None)

````python
def insertAt(head, given_val, idx):
    size = get_size(head)
    if idx < 0 or idx > size:
        print("Wrong Index. Insertion Failed!")
        return None

    newNode = Node(given_val)

    if idx == 0:# Special case for inserting at the beginning (head acts as the predecessor)
        predecessor = head
    else:
        # General case for inserting at idx (other than the start)
        predecessor = getNode(head, idx - 1)

    successor = predecessor.next  # Node currently at idx #

    # Establishing the connection of newNode
    newNode.next = successor
    newNode.prev = predecessor

    # Connecting rest of links
    predecessor.next = newNode
    if successor != None:  # Special case, succ exists, not None
        successor.prev = newNode
    # else--succesor is None--That means not a None--no connection


arr = [10, 20, 30, 40]
head3 = create_doubly_linear_LL(arr)
print("Original List:")
printLL_forward(head3)

# Insert 5 at index 0--HERE HEAD IS PRED
insertAt(head3, 5, 0)
print("\nList after inserting 5 at index 0 (just after head):")
printLL_forward(head3)


# Insert 15 at index 1
insertAt(head3, 15, 1)
print("\nList after inserting 15 at index 1 (middle):")
printLL_forward(head3)

# Insert 35 at index 3
insertAt(head3, 35, 3)
print("\nList after inserting 35 at index 3 (middle):")
printLL_forward(head3)

# Insert 50 at index 3
insertAt(head3, 50, 7)
print("\nList after inserting 50 at index 7 (after tail):")
printLL_forward(head3)
````

# 2.5.6 Removal:
- Removing an element from the list is done by removing the node that contains the element. Just like inserting a new node in a list, removing requires that you have the reference to the predecessor node.

- Since we're using a doubly-linked list, finding a predecessor of a node is trivial — it's n.prev.

- And thanks to the dummy head, there is no "special" case here as well.

## Illustration of Removal of any node after head to and before tail [Generic case]

## Illustration of Removal of tail (index = size-1)  (**Special case, Succesor is None**)

````python
def removeAt(head, idx):
    size = get_size(head)
    if idx < 0 or idx >= size:
        print("Wrong Index. Deletion Failed!")
        return

    remNode = getNode(head, idx)
    predecessor = remNode.prev
    successor = remNode.next

    # Connecting the rest of the links
    predecessor.next = successor
    if successor != None: # Special case--Succesor--None
        successor.prev = predecessor

    # Removing the connections of remNode
    remNode.next = remNode.prev = None


arr = [10, 20, 30, 40, 50]
head3 = create_doubly_linear_LL(arr)
print("Original List:")
printLL_forward(head3)

# Remove node at index 0
removeAt(head3, 0)
print("\nList after removing node at index 0 (Just after head):")
printLL_forward(head3)
print('-'*50)

# Remove node at index 2
removeAt(head3, 2)
print("\nList after removing node at index 2 (middle):")
printLL_forward(head3)
print('-'*50)

# Remove node at last index (now index 2 after removals)
removeAt(head3, 2)
print("\nList after removing node at last index 2 (tail):")
printLL_forward(head3)
print('-'*50)


# Remove node at wrong index
removeAt(head3, 5)
print("\nList after removing node at 5 (wrong index)")
printLL_forward(head3)
print('-'*50)

# Remove node at wrong index
removeAt(head3, -3)
print("\nList after removing node at -3 (wrong index)")
printLL_forward(head3)
print('-'*50)
````

## Reason for Doubly Linked List
By now, you know the difference between arrays and linked lists and their relative advantages and disadvantages. You learned about two types of linked list: 1) single-linked list and 2) double-linked list. Here we will discuss why double linked lists are important.

Although a **single-linked** list allows us to handle a linear sequence of elements that can be of unknown length and dynamic in our program, there are some **limitations in the traversal** of the list.

- **Traversal cost:**
    - Suppose we want to do read/remove/insert near at the end of the list. Then we have to traverse all the way from the head to our position of interest then do the desired operation. When the list is long then this traversal cost is a huge problem.
    - we have to be very careful regarding where to stop the traversal. For example, if you want to insert at position n, then we have to stop traversal at position n – 1. If you want to read then we stop traversal at n.

- **Cannot go backward/ no backtrack:** If we make one more wrong step then we cannot backtrack. You have to restart from the head.This second problem of not being able to move back-and-forth during our traversal is a major concern.

There are other variations of these linked lists as well. For example, Singly Circular Linked List, Doubly Linked List, Dummy Headed Doubly Linked List. You can try out these variations on your own.
```**Keep in mind**, if you properly understand how to connect and disconnect nodes, then you should be able to solve problems for any types of linked list```



#  Dummy-headed doubly linear Linked List Class - optional
Let's try to combine all the codes above and try to create a DHDLLL(Dummy headed doubly linear Linked List) Class

````python
# Defining the Node class
class Node:
    def __init__(self, elem, next=None, prev=None):
        self.elem = elem  # The element to store in the node
        self.next = next  # The reference to the next node
        self.prev = prev  # The reference to the previous node

#____________________________________________________________________
# Defining the DHDCLL class
class DHDLLL:
    def __init__(self):
        self.head = Node(None)  # Initialize dummy head node

    # Method to create a doubly linear linked list from an array
    def create_doubly_linear_LL(self, arr):
        current_node = self.head
        for i in range(len(arr)):
            newNode = Node(arr[i])
            current_node.next = newNode
            newNode.prev = current_node
            current_node = newNode
        return self.head  # Returning the head of the newly created linked list

    # Method to print the linked list in forward direction
    def print_forward(self):
        print("Forward Printing:", end=" ")
        current_node = self.head.next  # Start from the first real element
        while current_node:
            print(current_node.elem, end=" -> ")
            current_node = current_node.next
        print("END")  # End of the list

    # Method to print the linked list in backward direction
    def print_backward(self):
        print("Backward Printing:", end=" ")
        current_node = self.head
        # Traverse to the tail node first
        while current_node.next != None:
            current_node = current_node.next

        # Now traverse backwards
        while current_node.prev != None:
            print(current_node.elem, end=" <- ")
            current_node = current_node.prev
        print("HEAD")  # Indicating the start of the list

    # Method to get the size of the list
    def get_size(self):
        if self.head.next == None:  # If the list is empty
            return 0

        current_node = self.head.next  # Start from the first element after head
        count = 0
        while current_node != None:
            count += 1
            current_node = current_node.next
        return count

    # Method to get the node at a specific index
    def node_at(self, idx):
        current_node = self.head.next  # Start from the first real element
        count = 0
        while current_node != None:
            if count == idx:
                return current_node  # Return the node at the given index
            count += 1
            current_node = current_node.next
        return None  # Return None if index is out of bounds

    # Method to insert a value at a specific index
    def insert_at(self, given_val, idx):
        size = self.get_size()
        if idx < 0 or idx > size:
            print("Wrong Index. Insertion Failed!")
            return None

        newNode = Node(given_val)

        # Special case for inserting at the beginning (head as predecessor)
        if idx == 0:
            predecessor = self.head
        else:
            predecessor = self.node_at(idx - 1)

        successor = predecessor.next

        # Establishing the connection of newNode
        newNode.next = successor
        newNode.prev = predecessor

        # Connecting rest of links
        predecessor.next = newNode
        if successor != None:
            successor.prev = newNode

        return newNode

    # Method to remove a node at a specific index
    def remove_at(self, idx):
        size = self.get_size()
        if idx < 0 or idx >= size:
            print("Wrong Index. Deletion Failed!")
            return

        remNode = self.node_at(idx)
        predecessor = remNode.prev
        successor = remNode.next

        # Connecting the rest of the links
        predecessor.next = successor
        if successor!= None:
            successor.prev = predecessor

        # Removing the connections of remNode
        remNode.next = remNode.prev = None
````

````python
# Tester part

# Step 1: Create an instance of DHDLLL
linked_list = DHDLLL()

# Step 2: Initialize the list with an array
arr = [10, 20, 30, 40]
linked_list.create_doubly_linear_LL(arr)

# Step 3: Get the size of LL
print("Size of the linked list:", linked_list.get_size())  # Expected output: 4
print()

# Step 4: Try node_at()
index_to_find = 2
node2 = linked_list.node_at(index_to_find)
print("Node at index 2:", node2.elem if node2 else "None")  # Expected output: 30

index_to_find = 4
node3 = linked_list.node_at(index_to_find)
print("Node at index 4 (wrong index):", node3)  # Expected output: None
print()

# Step 5: Print the original list
print("Original List:")
linked_list.print_forward()
linked_list.print_backward()

# Step 6: Insert 15 at index 1
print("\nInserting 15 at index 1:")
linked_list.insert_at(15, 1)
linked_list.print_forward()

# Step 7: Insert 5 at index 0
print("\nInserting 5 at index 0:")
linked_list.insert_at(5, 0)
linked_list.print_forward()

# Step 8: Insert 45 at the end (index 6 after additions)
print("\nInserting 45 at the end:")
linked_list.insert_at(45, linked_list.get_size())
linked_list.print_forward()

# Step 9: Remove node at index 3
print("\nRemoving node at index 3:")
linked_list.remove_at(3)
linked_list.print_forward()

# Step 10: Remove node at index 0
print("\nRemoving node at index 0:")
linked_list.remove_at(0)
linked_list.print_forward()
````
